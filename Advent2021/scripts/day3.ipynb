```python
# from collections import Counter
# print(''.join([Counter(z).most_common(1)[0][0] for z in zip(*bits)]))
# a5cd

def parse(bits):
    word_list = []
    for word in bits:
        parsed_word = [i for i in word]
        parsed_word.pop(-1)
        word_list.append(parsed_word)
    return word_list
    
    
def most_least_common(word_list):
    bit_length = min([len(p) for p in word_list])
    most_common = ''
    least_common = ''
    for i in range(bit_length):
        bits = [int(word[i]) for word in word_list]
#         print(bits)
#         print(bits.dtype())
        a = mean(bits)
        mode = '1' if a >= .5 else '0'
        anti = '1' if a < .5 else '0'
    #     a = mode([int(word[i]) for word in word_list])
#         mode = str(a)
#         anti = str(1 - int(a))
        most_common += mode
        least_common += anti
    return most_common, least_common

def convert_to_decimal(bits):
    dec = int(bits, 2)
    return dec

    
word_list = parse(bits)
most, least = most_least_common(word_list)
print(most)
print(least)
print(convert_to_decimal(most) * convert_to_decimal(least))
```

```python
def filter_list(word_list, check_bit, most_common=True):
    most, least = most_least_common(word_list)
    valid_bit = most[check_bit] if most_common else least[check_bit]
    og_len = len(word_list)
    print(f"Filtering all words with bit {check_bit} != {valid_bit}, current length: {og_len}")
    filtered_list = list(filter(lambda word: word[check_bit] == valid_bit, word_list))
    print(f"Filtered {og_len - len(filtered_list)} elements, new length {len(filtered_list)}")
    return(filtered_list)

# l = filter_list(word_list, 0)

def recursively_filter(word_list, check_bit, most_common=True):
    if len(word_list) == 1 or  check_bit == 12:
        chosen = ''.join(word_list[0])
        return chosen, convert_to_decimal(chosen)
    else:
        word_list = filter_list(word_list, check_bit, most_common)
        check_bit += 1 
        return recursively_filter(word_list, check_bit, most_common)

most, least = most_least_common(word_list)
print(most)
print(least)

o2_gen = recursively_filter(word_list, 0)
print(f"o2 gen value is: {o2_gen} \n\n")
co2_scrub = recursively_filter(word_list, 0, False)
print(f"o2 scrub value is: {co2_scrub} \n\n")
print(o2_gen[1] *  co2_scrub[1])

```
