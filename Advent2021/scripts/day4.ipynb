```python
with open(f"Advent2021/inputs/input4.txt") as f:
    bingo_input  = f.readlines()

def strip_rows(bingo_input):
    return [p.strip() for p in bingo_input]

bingo_input = strip_rows(bingo_input)

def parse_read_numbers(bingo_input):
    numbers_read = bingo_input[0]
    numbers = numbers_read.split(',')
    return [int(i) for i in numbers]

def create_ind_boards(board_rows):
    board_rows = [row.split() for row in board_rows]
    board_rows = [list(map(int, row)) for row in board_rows]
    board_rows = np.array(board_rows)
    hits = np.zeros_like(board_rows)
    full_board = np.stack([board_rows, hits], axis=0)
#     print(full_board)
    return full_board
    
def parse_bingo_boards(bingo_input):
    bingo_input = bingo_input[2:]
    board_rows = []
    boards = []
    for row in bingo_input:
        if row == '':
            board = create_ind_boards(board_rows)
            board_rows = []
            boards.append(board)
        else:
            board_rows.append(row)
    return boards

numbers = parse_read_numbers(bingo_input)
boards = parse_bingo_boards(bingo_input)
# boards
# numbers
# # print(boards[0])
# boards[0][0][0][0] = 34
# print(boards[0][0])
# print()
# arr1 = boards[0][0] == 34
# arr2 = boards[0][1] == 1
# arr = np.logical_or(arr1, arr2)
# boards[0][1] = np.where(arr, 1, 0)
# print(boards[0])

# arr1 = boards[0][0] == 62
# arr2 = boards[0][1] == 1
# arr = np.logical_or(arr1, arr2)
# boards[0][1] = np.where(arr, 1, 0)
# print(boards[0])

# np.sum(boards[0][1], axis=1).max()

# t = np.where(boards[0][0] == 34)
# t = [list(p) for p in t]
# print(t)
# for z in zip(t[0], t[1]):
#     print(z)
# 
```

```python
def initialize_fresh_boards(boards):
    new_list = []
    for board in boards:
        new_list.append(np.copy(board))
    return new_list

def run_bingo_game(numbers, boards):
    fresh_copy = initialize_fresh_boards(boards)
    for number in numbers:
        print(f'We rolled .... {number} !!')
        bingo = check_all_boards(number, fresh_copy)
        if len(bingo) > 0: 
            break
    d = {"bingo": bingo[0], 'number': number}
    return  d

def check_all_boards(number, boards, excluded=[]):
#     print(len(boards))
    best = 0
    winners = []
    for i, board in enumerate(boards):
        if i in excluded:
            pass
        else:
            check_board_and_update(board, number)
            if is_bingo(board):
                print('BINGO!!')
                print(f"Board: {i}")
                winners.append((i, board))
    return winners
    
def check_board_and_update(board, number):
    hits = board[0] == number
    old_hits = board[1] == 1
    arr = np.logical_or(hits, old_hits)
    board[1] = np.where(arr, 1, 0)

def is_bingo(board):
    hit_array = board[1]
    rows = np.sum(hit_array, axis=0).max()
    columns = np.sum(hit_array, axis=1).max()
    if rows == 5 or columns == 5 :
        return True
    else:
        return False
    
d = run_bingo_game(numbers, boards)

def compute_bingo_score(board, final_number):
    print(board)
    logical_arr = board[1] == 0
    remaining_sum = np.where(logical_arr, board[0], 0).sum()
    print(remaining_sum * final_number)
    
print(d['bingo'])
compute_bingo_score(d['bingo'][1], d['number'])
```
